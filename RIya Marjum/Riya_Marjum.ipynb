```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
```

```python
np.random.seed(42)

data = {
    "Age": np.random.randint(18, 25, 100),
    "Study_Hours": np.random.uniform(1, 8, 100).round(1),
    "Attendance": np.random.randint(50, 100, 100),
    "Internal_Marks": np.random.randint(30, 100, 100)
}

df = pd.DataFrame(data)

df["Result"] = np.where(
    (df["Study_Hours"] >= 3) &
    (df["Attendance"] >= 75) &
    (df["Internal_Marks"] >= 40),
    "Pass", "Fail"
)

df.to_csv("student_performance.csv", index=False)

df.head()
```

```python
df = pd.read_csv("student_performance.csv")

print("Number of records:", df.shape[0])
print("Number of attributes:", df.shape[1])

print("\nClass Distribution:")
print(df["Result"].value_counts())
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
df["Result"] = df["Result"].map({"Pass": 1, "Fail": 0})
X = df.drop("Result", axis=1)
y = df["Result"]

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
temp_model = DecisionTreeClassifier(criterion="gini")
temp_model.fit(X_scaled, y)

print("Feature Importance (Gini Index):")
for feature, importance in zip(X.columns, temp_model.feature_importances_):
    print(feature, ":", round(importance, 3))
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

model = DecisionTreeClassifier(
    criterion="gini",
    max_depth=4,
    random_state=42
)

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
